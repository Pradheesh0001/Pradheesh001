```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import confusion_matrix


sentiment_data = pd.read_csv(r'C:\Users\Auditics Lap-\Downloads\sentiment_train.csv',encoding = 'unicode_escape')
sentiment_data.info()
```

```python
sentiment_data.isna().sum()
```

```python
sentiment_data= sentiment_data.fillna('ok')
```

```python
sentiment_data.isna().sum()
```

```python


x = sentiment_data['selected_text']
y = sentiment_data['sentiment']

print(x.shape)
print(y.shape)

```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3)

model = make_pipeline(TfidfVectorizer(),MultinomialNB())
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
cm = confusion_matrix(y_test,y_pred)

print(cm)


```

```python
pd.DataFrame(confusion_matrix(y_test,y_pred),columns = ['Predicted Negative','Predicted Neutral','Predicted Positive'],index = ['Actual Negative','Actual Neutral','Actual Positive'])
```

```python
from sklearn.metrics import accuracy_score

accuracy_info = accuracy_score(y_test,y_pred)
accuracy_info
```

```python
def predict_sentiment(txt):
    # Directly use the input text without cleaning
    pred = model.predict([txt])  # The model expects a list of inputs
    return pred
```

```python
predict_sentiment("I love Python")
```
