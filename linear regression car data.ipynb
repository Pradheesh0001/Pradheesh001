```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection  import train_test_split

car_data= pd.read_csv(r"C:\Users\Auditics Lap-\Downloads\CarPrice_Assignment.csv")
car_data.head()
```

```python
input_file = car_data.drop(columns = "price")
output_file = car_data["price"]

input_file.head()
```

```python
car_risk = car_data["symboling"]
fuel_type = car_data["fueltype"]
body_type = car_data["carbody"]

style.use('seaborn-v0_8-ticks')
chart = plt.figure(figsize = (40,30))

chart.suptitle("Car Price Ananalizer")

plt.subplot(1,3,1)
plt.bar(car_risk,output_file,color = ('Red','Yellow','Green'))
plt.xlabel("symboling")
plt.ylabel("price")

plt.subplot(1,3,2)
plt.bar(fuel_type,output_file, color = 'Green')
plt.xlabel("Fuel Type")
plt.ylabel("price")

plt.subplot(1,3,3)
plt.bar(body_type,output_file, color = ('Orange','blue','Green'))
plt.xlabel("Body Type")
plt.ylabel("Price")
```
